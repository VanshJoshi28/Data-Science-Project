## 🔧 3. Feature Engineering & Preprocessing

````python
# ── Select & clean features ───────────────────────────────────────────────────
FEATURES = ['iyear','imonth','region','attacktype1','targtype1',
            'weaptype1','suicide','multiple','nkill','nwound']
TARGET_BIN = 'success'          # binary  (Models 1-3)
TARGET_ATK = 'attacktype1'      # multiclass (Model 4)
TARGET_SEV = 'severity'         # engineered (Model 5)

df_clean = df[FEATURES + [TARGET_BIN]].copy()
df_clean['nkill']  = df_clean['nkill'].fillna(0)
df_clean['nwound'] = df_clean['nwound'].fillna(0)
df_clean = df_clean.dropna(subset=[TARGET_BIN])
df_clean = df_clean.fillna(df_clean.median(numeric_only=True))

# Engineered feature: severity label (0=low, 1=medium, 2=high)
df_clean['severity'] = pd.cut(
    df_clean['nkill'] + df_clean['nwound'],
    bins=[-1, 0, 5, 99999],
    labels=[0, 1, 2]
).astype(int)

print(f'✅ Clean dataset shape: {df_clean.shape}')
print(f'\nTarget distribution (success):\n{df_clean[TARGET_BIN].value_counts()}')
print(f'\nSeverity distribution:\n{df_clean["severity"].value_counts().sort_index()}')
df_clean.head(3)
````

````python
# ── Train/test split for binary classification ────────────────────────────────
X = df_clean[FEATURES]
y_bin = df_clean[TARGET_BIN]
y_atk = df_clean[TARGET_ATK]
y_sev = df_clean[TARGET_SEV]

X_tr, X_te, y_tr, y_te = train_test_split(X, y_bin, test_size=0.2,
                                           random_state=42, stratify=y_bin)

scaler = StandardScaler()
X_tr_sc = scaler.fit_transform(X_tr)
X_te_sc = scaler.transform(X_te)

# For multiclass (attack type), keep top 5 classes only to keep things clean
top5_atk = y_atk.value_counts().head(5).index
mask = y_atk.isin(top5_atk)
X_atk = X[mask]; y_atk = y_atk[mask]
X_atk_tr, X_atk_te, y_atk_tr, y_atk_te = train_test_split(
    X_atk, y_atk, test_size=0.2, random_state=42, stratify=y_atk)

# For KNN severity
X_sev_tr, X_sev_te, y_sev_tr, y_sev_te = train_test_split(
    X, y_sev, test_size=0.2, random_state=42, stratify=y_sev)
X_sev_tr_sc = scaler.fit_transform(X_sev_tr)
X_sev_te_sc = scaler.transform(X_sev_te)

print('✅ Splits ready.')
print(f'  Binary  — Train: {X_tr.shape}, Test: {X_te.shape}')
print(f'  AttType — Train: {X_atk_tr.shape}, Test: {X_atk_te.shape}')
print(f'  Severity— Train: {X_sev_tr.shape}, Test: {X_sev_te.shape}')
````

---
## 🔵 Model 1 — Logistic Regression
**Task:** Predict whether a terrorist attack will be **successful** (binary classification)

````python
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_tr_sc, y_tr)
y_pred_lr = lr.predict(X_te_sc)
y_prob_lr = lr.predict_proba(X_te_sc)[:,1]

acc_lr  = accuracy_score(y_te, y_pred_lr)
auc_lr  = roc_auc_score(y_te, y_prob_lr)
f1_lr   = f1_score(y_te, y_pred_lr)

print('='*52)
print('  MODEL 1 — LOGISTIC REGRESSION')
print('='*52)
print(f'  Accuracy : {acc_lr:.4f} ({acc_lr*100:.2f}%)')
print(f'  ROC-AUC  : {auc_lr:.4f}')
print(f'  F1-Score : {f1_lr:.4f}')
print('='*52)
print(classification_report(y_te, y_pred_lr, target_names=['Failed','Successful']))
````

````python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('🔵 Model 1 — Logistic Regression Results', fontsize=15, fontweight='bold')

# Confusion matrix
cm = confusion_matrix(y_te, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Failed','Success'], yticklabels=['Failed','Success'])
axes[0].set_title('Confusion Matrix')
axes[0].set_ylabel('Actual'); axes[0].set_xlabel('Predicted')

# ROC curve
fpr, tpr, _ = roc_curve(y_te, y_prob_lr)
axes[1].plot(fpr, tpr, color=PALETTE[1], lw=2, label=f'AUC = {auc_lr:.3f}')
axes[1].plot([0,1],[0,1],'--', color='gray', lw=1)
axes[1].fill_between(fpr, tpr, alpha=0.15, color=PALETTE[1])
axes[1].set_title('ROC Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

# Feature coefficients
coef = pd.Series(np.abs(lr.coef_[0]), index=FEATURES).sort_values(ascending=True)
coef.plot(kind='barh', ax=axes[2], color=PALETTE[2])
axes[2].set_title('Feature Importance (|Coefficient|)')
axes[2].set_xlabel('|Coefficient|')

plt.tight_layout()
plt.savefig('model1_lr.png', bbox_inches='tight')
plt.show()
````

````python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('🟢 Model 2 — Random Forest Results', fontsize=15, fontweight='bold')

# Confusion matrix
cm = confusion_matrix(y_te, y_pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', ax=axes[0],
            xticklabels=['Failed','Success'], yticklabels=['Failed','Success'])
axes[0].set_title('Confusion Matrix')
axes[0].set_ylabel('Actual'); axes[0].set_xlabel('Predicted')

# ROC curve
fpr, tpr, _ = roc_curve(y_te, y_prob_rf)
axes[1].plot(fpr, tpr, color=PALETTE[2], lw=2, label=f'AUC = {auc_rf:.3f}')
axes[1].plot([0,1],[0,1],'--', color='gray')
axes[1].fill_between(fpr, tpr, alpha=0.15, color=PALETTE[2])
axes[1].set_title('ROC Curve')
axes[1].set_xlabel('False Positive Rate'); axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

# Feature importance
fi = pd.Series(rf.feature_importances_, index=FEATURES).sort_values()
colors = [PALETTE[2] if v >= fi.quantile(0.5) else '#b0c4b1' for v in fi]
fi.plot(kind='barh', ax=axes[2], color=colors)
axes[2].set_title('Feature Importances')
axes[2].set_xlabel('Importance Score')

plt.tight_layout()
plt.savefig('model2_rf.png', bbox_inches='tight')
plt.show()
````

## 🔴 Model 4 — Decision Tree Classifier
**Task:** Classify the **attack type** (top 5 categories — multiclass classification)

````python
dt = DecisionTreeClassifier(max_depth=8, min_samples_leaf=50, random_state=42)
dt.fit(X_atk_tr, y_atk_tr)
y_pred_dt = dt.predict(X_atk_te)

acc_dt = accuracy_score(y_atk_te, y_pred_dt)
f1_dt  = f1_score(y_atk_te, y_pred_dt, average='weighted')

# Map code → readable label
atk_map = {
    1:'Assassination', 2:'Hijacking', 3:'Kidnapping',
    4:'Barricade Incident', 5:'Bombing/Explosion',
    6:'Armed Assault', 7:'Unarmed Assault', 8:'Facility/Infrastructure',
    9:'Unknown'
}
labels = [atk_map.get(c, str(c)) for c in sorted(y_atk_te.unique())]

print('='*52)
print('  MODEL 4 — DECISION TREE (Attack Type)')
print('='*52)
print(f'  Accuracy (weighted) : {acc_dt:.4f} ({acc_dt*100:.2f}%)')
print(f'  F1-Score (weighted) : {f1_dt:.4f}')
print('='*52)
print(classification_report(y_atk_te, y_pred_dt, target_names=labels))
````

## 🟣 Model 5 — K-Nearest Neighbors (KNN)
**Task:** Predict **casualty severity** (Low / Medium / High) based on incident features

````python
# Find optimal K
k_range = range(3, 22, 2)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(X_sev_tr_sc, y_sev_tr)
    k_scores.append(accuracy_score(y_sev_te, knn.predict(X_sev_te_sc)))

best_k = list(k_range)[k_scores.index(max(k_scores))]
print(f'🔍 Optimal K = {best_k}  (accuracy = {max(k_scores):.4f})')

knn = KNeighborsClassifier(n_neighbors=best_k, n_jobs=-1)
knn.fit(X_sev_tr_sc, y_sev_tr)
y_pred_knn = knn.predict(X_sev_te_sc)

acc_knn = accuracy_score(y_sev_te, y_pred_knn)
f1_knn  = f1_score(y_sev_te, y_pred_knn, average='weighted')

print('='*52)
print('  MODEL 5 — K-NEAREST NEIGHBORS (KNN)')
print('='*52)
print(f'  Best K   : {best_k}')
print(f'  Accuracy : {acc_knn:.4f} ({acc_knn*100:.2f}%)')
print(f'  F1-Score : {f1_knn:.4f}')
print('='*52)
print(classification_report(y_sev_te, y_pred_knn,
      target_names=['Low','Medium','High']))
````

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('🟣 Model 5 — KNN Results', fontsize=15, fontweight='bold')

# K vs Accuracy
axes[0].plot(list(k_range), k_scores, 'o-', color=PALETTE[4], linewidth=2, markersize=7)
axes[0].axvline(best_k, color=PALETTE[0], linestyle='--', label=f'Best K={best_k}')
axes[0].set_title('K vs Accuracy')
axes[0].set_xlabel('K (number of neighbors)')
axes[0].set_ylabel('Accuracy')
axes[0].legend()

# Confusion matrix
cm = confusion_matrix(y_sev_te, y_pred_knn)
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples', ax=axes[1],
            xticklabels=['Low','Medium','High'],
            yticklabels=['Low','Medium','High'])
axes[1].set_title('Confusion Matrix — Severity')
axes[1].set_ylabel('Actual'); axes[1].set_xlabel('Predicted')

plt.tight_layout()
plt.savefig('model5_knn.png', bbox_inches='tight')
plt.show()
````
